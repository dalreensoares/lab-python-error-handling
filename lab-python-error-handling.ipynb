def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Initialize inventory with user input
def initialize_inventory(products):
    inventory = {}
    for product in products:
        while True:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                inventory[product] = quantity
                break
            except ValueError as error:
                print(f"Error: {error}")
    return inventory





````

````python
#2.

def total_products_ordered(customer_order):
    prices = []
    for product_name in customer_order:
        while True:
            try:
                price = float(input(f"The price of {product_name} is: "))
                if price < 0:
                    raise ValueError("The price cannot be negative.")
                prices.append(price)
                break
            except ValueError as e:
                print(f"Invalid input: {e}. Please enter a valid price for {product_name}.")
    return prices

def calculate_total_price(prices):
    return sum(prices)
````

````python
#3.
def get_customer_orders(products, inventory):
    customer_order = []
    while True:
        try:
            num_orders = int(input("How many different products would you like to order? "))
            if num_orders <= 0:
                raise ValueError("The number of orders must be a positive integer.")
            break
        except ValueError as e:
            print(f"Invalid input: {e}. Please enter a valid number of orders.")
    
    for _ in range(num_orders):
        while True:
            try:
                product_name = input("Enter the product name: ").strip().lower()
                if product_name not in products:
                    raise ValueError("The product is not in the inventory.")
                if inventory[product_name] <= 0:
                    raise ValueError("The product is out of stock.")
                customer_order.append(product_name)
                break
            except ValueError as e:
                print(f"Invalid input: {e}. Please enter a valid product name that is in stock.")
    
    return customer_order
````

````python
#4:
inventory = initialize_inventory(products)
customer_order = get_customer_orders(products, inventory)
ordered_prices = total_products_ordered(customer_order)
total_price = calculate_total_price(ordered_prices)
print(f"Customer order: {customer_order}")
print(f"The total price of the ordered products is: {total_price:.2f}")
````
